# EXPLORATORY DATA ANALYSIS

```python
countries_with_most_terrorism = data.country_txt.value_counts().head(10)
countries = list(countries_with_most_terrorism.index)
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(countries_with_most_terrorism.index,countries_with_most_terrorism.values)
plt.title('Countries with most terrorism')
```

Iraq is the country with most Terrorism 
Turkey has lowest Terrorism

```python
year = data.iyear.value_counts().head(5)
plt.bar(year.index,year.values)
plt.title("Terrorist activity in the recent Years")
```

In 2014, terror activity was the heighest .but in recent years terror activities is decreasing

**Type of attack**

```python
plt.subplots(figsize=(15,6))
sns.countplot('attacktype1_txt',data=data,palette='rocket',order=data['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attacking methods by terrorists')
plt.show()
```

Most of the terrorist are using bombing/explosion and  armed assualt. 

**Citywise attack** 

```python


plt.subplots(figsize=(18,6))
sns.barplot(data['city'].value_counts()[1:15].index,data['city'].value_counts()[1:15].values)
plt.show()
```

The most affected cities are Baghdad  and Karachi 

**Province wise terror attack** 

```python
from wordcloud import WordCloud
from scipy import signal
cities = data.provstate.dropna(False)
plt.subplots(figsize=(20,20))
wordcloud=WordCloud(background_color='black',width=512,height=382).generate(' '.join(cities))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

This plot represents which state are most affected . The more the size of the content  more is the terrorism

**Favorite targets**

```python
plt.subplots(figsize=(15,6))
sns.countplot('targtype1_txt',data=data,palette='mako',order=data['targtype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Favorite targets")
plt.show()
```

the favourite target of the terrorist is Private citizens and property and Military . 

**Countries with most terrorism on religious figures**

```python
plt.bar(data['country_txt'].value_counts().head().index,data['country_txt'].value_counts().head().values)
plt.title("Countries with most terrorism on religious figures")
```

Countries with most terrorism on religious figures is Iraq

**Region wise Terror attacks**

```python
sns.countplot('region_txt',data=data,palette='Paired',edgecolor=sns.color_palette('dark',7),order=data['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
terror_region=pd.crosstab(data.iyear,data.region_txt)
terror_region.plot(color=sns.color_palette('Paired',12))
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show
```
